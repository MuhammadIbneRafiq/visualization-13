```python
def list_attributes_with_high_missing_values(dataframe, threshold=0.95):
    """
    Lists attributes (columns) with missing values greater than the specified threshold.
    Parameters:
        dataframe (pd.DataFrame): The input DataFrame.
        threshold (float): The fraction of missing values to consider (default is 0.95).
    Returns:
        list: A list of column names with more than the threshold percentage of missing values.
    """
    # Calculate the fraction of missing values for each column
    missing_values = dataframe.isnull().mean()
    
    # Filter columns where the missing fraction is greater than the threshold
    high_missing_columns = missing_values[missing_values > threshold].index.tolist()
    
    return high_missing_columns

high_missing_attributes = list_attributes_with_high_missing_values(df_shark)

high_missing_attributes
```

```python
# Drop the attributes with high missing values
df_shark = df_shark.drop(columns=high_missing_attributes)

df_shark.sample(5)
    
```
